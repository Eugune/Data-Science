# Market segmentation example

## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

```python
df = pd.read_csv('3.12.+Example.csv')
```

```python
df.head()
```

## Plot the data

```python
plt.scatter(df.Satisfaction,df.Loyalty)
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

## Select the features

```python
x = df.copy()
```

## Clustering

```python
kmeans= KMeans(2)
kmeans.fit(x)
```

## Clustering results

```python
identified_cluster = kmeans.fit_predict(x)
```

```python
x['cluster'] = identified_cluster
plt.scatter(x.Satisfaction,x.Loyalty, c=x.cluster, cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

## Standardize the variables

```python
from sklearn import preprocessing

x_scaled = preprocessing.scale(x)
x_scaled
```

## Take advantage of the Elbow method

```python
wcss = []

for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(1,10), wcss)
plt.xlabel('numbers of cluster')
plt.ylabel('within cluster sum of square')
plt.show()
```

```python
# As you see from the graph, to get the best resut of clustering, the number of cluster should between [2,5] with standardization.
```
