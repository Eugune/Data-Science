## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Countries-exercise.csv'</i>.

```python
df_raw = pd.read_csv('Countries-exercise.csv')
```

```python
df_raw.head()
```

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
sns.scatterplot(x='Longitude',y='Latitude', data = df_raw)
plt.xlim(-180,180)
plt.ylim(-90, 90)
plt.show()
```

## Select the features

Make sure to select the appropriate features since we are no longer using the categorical variable for our clustering but rather <i>'Longitude'</i> and <i>'Laditude'</i>.

```python
x = df_raw.iloc[:,1:3]
```

## Clustering

Use 4 clusters initially.

```python
kmeans = KMeans(4)
kmeans.fit(x)
```

### Clustering results

```python
identified_cluster = kmeans.fit_predict(x)
```

```python
df_raw['cluster'] = identified_cluster
```

Plot the data once again but separate the data by the clusters we defined.  

## Plot the data

```python
sns.scatterplot(x='Longitude',y='Latitude', data = df_raw, c=df_raw.cluster, cmap='rainbow')
plt.xlim(-180,180)
plt.ylim(-90, 90)
plt.show()
```

## Selecting the number of clusters

### WCSS

Use the ingerated <i>sklearn</i> method <i> 'inertia_' </i>.

```python
kmeans.inertia_
```

```python
wscc=[]
cl_num = 11
for i in range(1,cl_num):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wscc.append(kmeans.inertia_)
```

Write a loop that calculates and saves the WCSS for any number of clusters from 1 up to 10 (or more if you wish).

```python
plt.plot(range(1,11), wscc)
plt.ylabel('Within-cluster Sum of Squares')
plt.xlabel('Number of clusters')
plt.show()
```
