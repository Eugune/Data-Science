## Import the relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
sns.set()
```

## Load the data

Load data from the csv file: <i> 'Countries.csv'</i>.


```python
df = pd.read_csv('Countries-exercise.csv')
```

```python
df.head()
```

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
plt.scatter(df.Longitude, df.Latitude)
plt.xlim(-180,180)
plt.ylim(-90, 90)
plt.show()
```

## Select the features

Create a copy of that data and remove all parameters apart from <i>Longitude</i> and <i>Latitude</i>.

```python
df_c = df.copy()
df_c.drop(df_c.columns[0], axis=1, inplace=True)
```

## Clustering

Assume there are only two clusters. 

```python
kmeans = KMeans(2)
```

```python
kmeans.fit(df_c)
```

```python
identified_clusters  = kmeans.fit_predict(df_c)
identified_clusters
```

### Clustering Resutls

```python
df_with_cluster = df.copy()
df_with_cluster['cluster'] = identified_clusters
df_with_cluster
```

```python
plt.scatter(df_with_cluster.Longitude, df_with_cluster.Latitude, c=df_with_cluster.cluster, cmap='rainbow')
plt.xlim(-180,180)
plt.ylim(-90, 90)
plt.show()
```

Did you remember to use the <i> c </i> parameter to separate the data by the clusters we defined?
